```python
import numpy as np
from matplotlib import pyplot as plt

fig = plt.figure(figsize = (10, 5))

def createf_accbycount(data, xn, yn, ti):
    xl = list(data.keys())
    yl = list(data.values())
    score = []
    for i in yl:
        try:
            score.append(i['accuracy']/i['count']*100)
        except:
            score.append(0)
    
    plt.bar(xl, score, width=0.4)
    plt.xlabel(xn)
    plt.ylabel(yn)
    plt.title(ti)
    plt.savefig(str(ti+".png"))
    plt.show()

def createf_acconly(data, xn, yn, ti):
    xl = list(data.keys())
    yl = list(data.values())
    score = []
    for i in yl:
        try:
            score.append(i['accuracy']*100)
        except:
            score.append(0)
    
    plt.bar(xl, score, width=0.4)
    plt.xlabel(xn)
    plt.ylabel(yn)
    plt.title(ti)
    plt.savefig(str(ti+".png"))
    plt.show()
    
    
# creating Part Analysis
parts_data = {'Bumper': {'tot_f1': 1.0, 'tot_fnr': 0.0, 'tot_fpr': 0.0, 'accuracy': 27.0, 'count': 27},
              'T': {'tot_f1': 1.0, 'tot_fnr': 0.0, 'tot_fpr': 0.0, 'accuracy': 5.0, 'count': 5},
              'F': {'tot_f1': 0.9111111111111111, 'tot_fnr': 0.1, 'tot_fpr': 0.0, 'accuracy': 13.504950495049505, 'count': 15},
              'D': {'tot_f1': 1.0, 'tot_fnr': 0.0, 'tot_fpr': 0.0, 'accuracy': 15.0, 'count': 15},
              'QP': {'tot_f1': 0.7333333333333333, 'tot_fnr': 0.26666666666666666, 'tot_fpr': 0.0, 'accuracy': 11.0, 'count': 15},
              'B': {'tot_f1': 1.0, 'tot_fnr': 0.0, 'tot_fpr': 0.0, 'accuracy': 14.0, 'count': 14}}
createf_accbycount(parts_data, "Parts", "Mask Quality", "Part Analysis")

#Creating Damages
damages_data = {'S': {'tot_f1': 0.0, 'tot_fnr': 1.0, 'tot_fpr': 0.0, 'accuracy': 0.0, 'count': 3}, 'MD': {'tot_f1': 0.14285714285714285, 'tot_fnr': 0.8571428571428571, 'tot_fpr': 0.0, 'accuracy': 0.14285714285714285, 'count': 7}}
createf_accbycount(damages_data, "Damage Type", "Mask Quality", "Damage Analysis")

rearview_data = {'Bumper': {'accuracy': 1.0, 'count': 15}, 'T': {'accuracy': 1.0, 'count': 5}, 'S': {'accuracy': 0, 'count': 0}, 'MD': {'accuracy': 0, 'count': 0}}
createf_acconly(rearview_data, "Parts/Damages", "Mask Quality", "Rear View")

rightview_data = {'F': {'accuracy': 0.9292786421499292, 'count': 7}, 'D': {'accuracy': 1.0, 'count': 7}, 'QP': {'accuracy': 0.5, 'count': 6}, 'S': {'accuracy': 0.0, 'count': 1}, 'MD': {'accuracy': 1.0, 'count': 1}}
createf_acconly(rightview_data, "Parts/Damages", "Mask Quality", "Right View")

leftview_data = {'F': {'accuracy': 0.875, 'count': 8}, 'D': {'accuracy': 1.0, 'count': 8}, 'QP': {'accuracy': 0.8888888888888888, 'count': 9}, 'S': {'accuracy': 0.0, 'count': 1}, 'MD': {'accuracy': 0.0, 'count': 2}}
createf_acconly(leftview_data, "Parts/Damages", "Mask Quality", "Left View")

frontview_data = {'Bumper': {'accuracy': 1.0, 'count': 12}, 'B': {'accuracy': 1.0, 'count': 14}, 'S': {'accuracy': 0.0, 'count': 1}, 'MD': {'accuracy': 0.0, 'count': 4}}
createf_acconly(frontview_data, "Parts/Damages", "Mask Quality", "Front View")
```
